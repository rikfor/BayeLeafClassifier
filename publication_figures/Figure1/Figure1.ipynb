```python
import anndata as ad
import scanpy as sc
import numpy as np
import pickle
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Show separation in example data for example gene

```python
data = sc.read('density_data_forFigure1B.h5ad')
```

```python
cd8ts = data[data.obs.CellType == 'CD8T'].to_df()
class_mono = data[data.obs.CellType == 'Classical Monocyte'].to_df()
```

```python
cd8ts = data[data.obs.CellType == 'CD8T'].to_df()
cd4ts = data[data.obs.CellType == 'CD4T'].to_df()
```

```python
gene = 'CD3D'
sns.kdeplot(cd8ts[gene], fill = True, bw_adjust=0.5, label = 'CD8T')
sns.kdeplot(cd4ts[gene], fill = True, bw_adjust=0.5, label = 'CD4T')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend()

#plt.savefig('figure1A_density.svg', dpi = 200, bbox_inches = 'tight')
```

```python
cd8ts = data[data.obs.CellType == 'CD8T'].to_df()
cms = data[data.obs.CellType == 'Classical Monocyte'].to_df()
```

```python
gene = 'CD3D'
sns.kdeplot(cd8ts[gene], fill = True, bw_adjust=0.5, label = 'CD8T')
sns.kdeplot(cms[gene], fill = True, bw_adjust=0.5, label = 'Classical Monocyte', color = 'tab:green')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend()

#plt.savefig('figure1Bdensity.svg', dpi = 200, bbox_inches = 'tight')
```

```python
gene = 'LYZ'
sns.kdeplot(cd8ts[gene], fill = True, bw_adjust=0.5, label = 'CD8T')
sns.kdeplot(cms[gene], fill = True, bw_adjust=0.5, label = 'Classical Monocyte', color = 'tab:green')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend()

#plt.savefig('figure1Cdensity.svg', dpi = 200, bbox_inches = 'tight')
```
